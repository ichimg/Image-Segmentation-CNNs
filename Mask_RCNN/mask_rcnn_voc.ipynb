```python
# Load the dataset_train from the file
output_file_path = '/content/drive/MyDrive/DL/ImageSegmentation/dataset_train.pkl'
with open(output_file_path, 'rb') as input_file:
    dataset_train = pickle.load(input_file)
dataset_train.prepare()

# Load the dataset_train from the file
output_file_path = '/content/drive/MyDrive/DL/ImageSegmentation/dataset_val.pkl'
with open(output_file_path, 'rb') as input_file:
    dataset_val = pickle.load(input_file)
dataset_val.prepare()
```

```python
config = VocConfig()
model = modellib.MaskRCNN(mode="training", config=config, model_dir='/content/drive/MyDrive/DL/ImageSegmentation')
```

```python
import os
ROOT_DIR = os.path.abspath("/content/drive/MyDrive/DL/ImageSegmentation/mask_rcnn")
# Path to trained weights file
COCO_WEIGHTS_PATH = os.path.join(ROOT_DIR, "mask_rcnn_coco.h5")

# Download weights file
if not os.path.exists(COCO_WEIGHTS_PATH):
    mrcnn.utils.download_trained_weights(COCO_WEIGHTS_PATH)

# Load weights trained on MS-COCO
model.load_weights(COCO_WEIGHTS_PATH, by_name=True, exclude=[
    "mrcnn_class_logits", "mrcnn_bbox_fc", "mrcnn_bbox", "mrcnn_mask"])
```

```python
print("Training network heads")
model.train(dataset_train, dataset_val, learning_rate=0.01, epochs=5, layers='heads')
```
